## Library imports + importing dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("IMDb_Dataset_Edited.csv")

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Sample of dataset

```python
df.sample(3)
```

# Dataset info with df.info()

```python
df.info()
```

```python
# Dropping all columns with significant (>50%) amount of null values (with the exception of IMDB rating)
df.drop(columns=["imdb_id","tagline","production_companies","production_countries","spoken_languages","AverageRating","Poster_Link","Certificate","Meta_score","Star1","Star2","Star3","Star4","Director_of_Photography","Producers","Music_Composer"], inplace=True)
df.info()
```

```python
df["IMDB_Rating"].fillna(-1, inplace=True) # fill missing rating valeus with 0
df.fillna("Missing",inplace=True) # Filling all remaining null values with "Missing"
df.info()
```

```python
df.nunique() # Counting the number of unique values per column
```

```python
df.drop_duplicates(inplace=True) # dropping duplicates while considering all columns
df.info()
```

```python
# Examining the values found under status and original language
df.value_counts(["status","original_language"])
```

```python
# Counting number of rows with 0 values for each numerical column
print("Number of entries with vote_average = 0 is ", df[df['vote_average'] == 0]['title'].count())
print("Number of entries with vote_count = 0 is ",  df[df['vote_count'] == 0]['title'].count())
print("Number of entries with revenue = 0 is ", df[df['revenue'] == 0]['title'].count())
print("Number of entries with runtime = 0 is ", df[df['runtime'] == 0]['title'].count())
print("Number of entries with budget = 0 is ", df[df['budget'] == 0]['title'].count())
print("Number of entries with popularity = 0 is ", df[df['popularity'] == 0]['title'].count())
```

```python
df.drop(columns=["vote_average","vote_count","revenue","budget", "keywords","overview_sentiment", "original_title"], inplace=True) # dropping numerical columns with significant amount of 0 values + keywords, overview_sentiment, original_title because it is unnecessary for this assignment
df.info()
```

```python
# Exporting processed dataset
df.to_csv("Processed IMDB Dataset.csv", index = False)
```
